**Realce de Imagens**

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img01 = cv.imread("/content/img01.jpg")
plt.imshow(img01, cmap="gray")
```

```python
img02 = cv.imread("/content/img02.jpg")
plt.imshow(img02, cmap="gray")
```

```python
img03 = cv.imread("/content/img03.jpg")
plt.imshow(img03, cmap="gray")
```

```python
img04 = cv.imread("/content/img04.jpg")
plt.imshow(img04)
```

```python
def get_histogram(imagem, limite_cinza):
    array = imagem.flatten()
    hist = np.zeros(limite_cinza,dtype='int')
    for i in range(0, len(array)):
        hist[array[i]] +=1
    return hist
```

```python
histograma = get_histogram(img01, 256)
print(histograma)
```

```python
histograma2 = get_histogram(img02, 256)
print(histograma2)
```

```python
histograma3 = get_histogram(img03, 256)
print(histograma3)
```

```python
histograma04 = get_histogram(img04, 256)
print(histograma04)
```

```python
plt.grid("on")

x = range(len(histograma))
y = histograma

plt.plot(x,y)
```

```python
plt.grid("on")

x = range(len(histograma2))
y = histograma2

plt.plot(x,y)
```

```python
plt.grid("on")

x = range(len(histograma3))
y = histograma3

plt.plot(x,y)
```

```python
plt.grid("on")

x = range(len(histograma04))
y = histograma04

plt.plot(x,y)
```

* Histograma com baixo contraste

```python
h = cv.calcHist([img01], [0], None, [256], [0,256])

plt.figure()
plt.title("Histograma Imagem Cinza")
plt.xlabel("Intensidade")
plt.ylabel("Quantidade de pixels")

plt.plot(h)
plt.xlim([0,256])
plt.show()
```

```python
h = cv.calcHist([img04], [0], None, [256], [0,256])

plt.figure()
plt.title("Histograma Imagem Colorida")
plt.xlabel("Intensidade")
plt.ylabel("Quantidade de pixels")

plt.plot(h)
plt.xlim([0,256])
plt.show()
```
