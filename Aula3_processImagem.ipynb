### Relacionamento entre elementos de uma imagem

* Vizinhança-4

```python
def vizinhanca4(linha, coluna):

  v1 = (linha - 1, coluna)
  v2 = (linha, coluna - 1)
  v3 = (linha + 1, coluna)
  v4 = (linha, coluna + 1)

  return [v1, v2, v3, v4]
```

* Vizinhança-8

```python
def vizinhanca8(linha, coluna):

  v5 = (linha - 1, coluna - 1)
  v6 = (linha - 1, coluna + 1)
  v7 = (linha + 1, coluna - 1)
  v8 = (linha + 1, coluna + 1)

  vizinhanca = vizinhanca4(linha, coluna) + [v5, v6, v7, v8]
  return vizinhanca
```

### Cortar imagens + Criar ruído + Retirar ruído

* Corte da imagem

```python
origem = (100, 100)
delta = 200

new_image = image[origem[0]:(origem[0] + delta), origem[1]:(origem[1] + delta)]

plt.imshow(new_image, cmap="gray")
plt.show() 
```

* Criando ruídos na imagem

```python
temp_image = np.asarray(image)
```

```python
tamanho_imagem = image.shape

M = tamanho_imagem[0]
N = tamanho_imagem[1]

new_image = np.zeros((M, N), dtype='uint8')

for i in range(0, M):
  randomValues = random.sample(range(0, N), 40)
  for value in randomValues:
    temp_image[i, value] = 0
```

```python
plt.imshow(temp_image, cmap="gray")
plt.show()
```

* Retirando os ruídos da imagem

```python
median = cv.medianBlur(temp_image, 5)
```

```python
plt.subplot(121), plt.imshow(temp_image), plt.title('Com ruído')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(median), plt.title('Sem ruído')
plt.xticks([]), plt.yticks([])

plt.show() 
```
