```python
img = data.astronaut()
```

```python
plt.imshow(img)
plt.show()
```

```python
from skimage.color import rgb2gray, rgb2gray

img_gray = rgb2gray(img)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```
