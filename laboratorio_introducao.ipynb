```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow 
from skimage import io
import matplotlib.pylab as plt
```

**Etapa 1:**

Ler uma imagem de url e exibi-lá usando openCV

```python
# carregando a imagem
imageurl = "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"

# lendo a imagem
image = io.imread(imageurl)

# exibindo a imagem
cv2_imshow(image) 
```

**Etapa 2:**

Visualizar a imagem

```python
print(image.shape)
```

```python
(h, w, c) = image.shape

print('Largura: {} pixels'.format(w))
print('Altura: {} pixels'.format(h))
print('Canais: {}'.format(c)) 
```

**Etapa 3:**

Conferir as cores do pixel

```python
print(image)
```

```python
print(image[0,0])
```

**Etapa 4:**

Usando escala de cinza para converter a imagem para os níveis de cinza

```python
from skimage.color import rgb2gray, rgb2gray

img_gray = rgb2gray(image)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

**Etapa 5:**

Exibindo o tamanho da imagem na escala de cinza 

```python
print(img_gray.shape) 
```

**Etapa 6:** 

Exibindo a matriz da imagem na escala de cinza

```python
print(img_gray)
```

**Etapa 7:**

Redimensionar e rotacionar uma imagem

```python
from skimage.transform import resize, rotate

img_r = image[:, :, 0]
plt.imshow(img_r)
plt.show()

img_g = image[:, :, 2]
plt.imshow(img_g)
plt.show()

print('Maximo de pixels: ')
print(img_g.max())

print('Minimo de pixels: ')
print(img_g.min())

print('Media de pixels: ')
print(img_g.mean()) 
```

```python
a, l, c = image.shape
img_red = resize(image, (int(a/2), int(l/2)))
plt.imshow(img_red)
plt.show() 
```

```python
a, l, c = image.shape
img_red2 = resize(image, (int(a*2), int(l*2)))
plt.imshow(img_red2)
plt.show() 
```

```python
img_rot = rotate(image, 90, resize=True)
plt.imshow(img_rot)
plt.show() 
```

```python
img_rot = rotate(image, -90, resize=True)
plt.imshow(img_rot)
plt.show() 
```

```python
img_rot = rotate(image, 180, resize=True)
plt.imshow(img_rot)
plt.show() 
```

```python
img_rot = rotate(image, -180, resize=True)
plt.imshow(img_rot)
plt.show() 
```

**Etapa 8:**

Exibir todos os pixels da imagem em um histograma

```python
plt.hist(image.ravel(), bins=256, range=[0,256])
plt.show()
```

**Etapa 10:**

Transformar uma imagem RGB em uma imagem de tons de cinza

```python
gray_image = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image) 
```

```python
plt.hist(gray_image.ravel(), bins=256, range=[0,256])
plt.show()
```

**Etapa 12:**

Transformação em tons de cinza (operação inversa)

```python
new_image1 = 255 - gray_image
cv2_imshow(new_image1) 
```

```python
new_image2 = (100.0/255)*gray_image + 100
cv2_imshow(new_image2) 
```

```python
new_image3 = 255.0 * (gray_image/255.0)**2
cv2_imshow(new_image3) 
```
